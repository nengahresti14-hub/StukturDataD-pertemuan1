membuat list


```python
list = []
```

membuat input 10 data dengan perulangan for

```python
for i in range(10):
    angka = int(input("Masukkan angka: "))
    list.append(angka)
```

membuat Tampilkan data list, nilai tertinggi, terendah dan Rata-rata

```python
print("List:", list)
print("Nilai maksimum:", max(list))
print("Nilai minimum:", min(list))
print("Rata-rata:", sum(list)/len(list))
```

buat grafik bar

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("Grafik Array")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()
```
